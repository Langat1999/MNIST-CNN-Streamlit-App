This code performs Named Entity Recognition (NER) and sentiment analysis on a list of product reviews using the spaCy and TextBlob libraries. It identifies entities like organizations (e.g., Samsung Galaxy, Sony) and determines the sentiment of each review (Positive or Negative).

```python
# ğŸ’¬ Task 3: Named Entity Recognition with spaCy

import spacy
from textblob import TextBlob

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Sample text data (you can add more)
reviews = [
    "I love the battery life of my new Samsung Galaxy!",
    "The Apple AirPods are too expensive for the quality.",
    "Sony headphones provide amazing sound clarity."
]

for text in reviews:
    doc = nlp(text)
    print(f"\nğŸ”¹ Review: {text}")
    print("Entities:", [(ent.text, ent.label_) for ent in doc.ents])

    sentiment = TextBlob(text).sentiment.polarity
    sentiment_label = "Positive" if sentiment > 0 else "Negative"
    print("Sentiment:", sentiment_label)
```
